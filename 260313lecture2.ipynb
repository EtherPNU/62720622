```python
# 사칙연산 프로그램(변수 + 연산자 + 출력문만 사용)

# 1) 숫자를 변수에 저장(값을 담아두는 상자라고 생각하면 돼요)
a = 10
b = 3

# 2) 덧셈: a와 b를 더한 결과를 add_result에 저장
add_result = a + b

# 3) 뺄셈: a에서 b를 뺀 결과를 sub_result에 저장
sub_result = a - b

# 4) 곱셈: a와 b를 곱한 결과를 mul_result에 저장
mul_result = a * b

# 5) 나눗셈: a를 b로 나눈 결과를 div_result에 저장
div_result = a / b

# 6) 출력: 변수에 들어있는 값(결과)을 화면에 보여주기
print("a =", a)
print("b =", b)

print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
```

```python

# 사칙연산 프로그램(입력 + 연산 + 출력)
# 사용자가 a, b 값을 직접 입력할 수 있게 만든 버전입니다.

# 주의: 변수 이름을 print, input 같은 내장 함수 이름으로 쓰면 오류가 날 수 있어요.

# 1) 사용자에게 숫자 2개 입력받기 (input은 문자열이므로 int/float로 바꿔줘야 함)
a = float(input("첫 번째 숫자 a를 입력하세요: "))
b = float(input("두 번째 숫자 b를 입력하세요: "))

# 2) 사칙연산 결과 계산
add_result = a + b
sub_result = a - b
mul_result = a * b

# 3) 나눗셈은 b가 0일 수 있으니 예외 처리
if b != 0:
    div_result = a / b
else:
    div_result = None  # 0으로 나눌 수 없으니 결과를 없다고 표시

# 4) 출력
print("a =", a)
print("b =", b)

print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)

if div_result is not None:
    print("a / b =", div_result)
else:
    print("a / b = (오류) 0으로 나눌 수 없습니다.")
```
