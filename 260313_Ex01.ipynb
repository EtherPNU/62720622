```python
a=10
b=3
add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = a / b
print("a =", a)
print("b =", b)

print("a + b =", add_result)
print("a - b =", sub_result)
print("a * b =", mul_result)
print("a / b =", div_result)
#오류가 생기는 이유는 위에 셀에서 변수 a와b를 지정 해놓았기 때문에 a와b의 변수 지정을 다시 초기화해 주어야함
```
