```python
height1 = float(input("학생1의 키를 입력하시오: "))
height2 = float(input("학생2의 키를 입력하시오: "))

if height1 == height2:
    print('두 학생의 키가 같다')
elif height1 > height2:  # else는 조건이 붙는 것이 아닌 if문에서 참이 아닐때 값을 되돌려 주는 것이므로 else에선 조건이 없음 elif는 갯수제한이없음
    print('학생1의 키가 더 크다')
else:
    print('학생2의 키가 더 크다')
```

```python
score = float(input('점수를 입력하세요')) #if문 예제1
if score >=90:
  print('A학점')
elif score >=80:
  print('B학점')
elif score >=70:
  print('C학점')
else:
  print('D학점')  #else문에서 조건이 필요가 없으므로 판단의 갯수를 n개라 가정할때 조건문은 n-1개이다.
```

```python
import random #if문 예제2

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f'주사위1:{dice1}')
print(f'주사위2:{dice2}')
if dice1 == dice2:
  print('두 주사위 값이 같다')
elif dice1 > dice2:
  print('주사위1의 값이 크다')
else:
  print('주사위2의 값이 크다')
```

```python
year = int(input('연도를 입력하시오'))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year}년은 윤년입니다.")
else:
    print(f"{year}년은 평년입니다.")
```

```python
#가위바위보
user1=random.randint(1,3)
user2=random.randint(1,3)
print(f'user1:{user1}')
print(f'user2:{user2}')

if user1 == user2:
    print('무승부')
elif (user1==1 and user2==2) or (user1==2 and user2==3) or (user1==3 and user2==1):
  print('유저1이 승리하였습니다')
else:
  print('유저2가 승리하였습니다')

```
