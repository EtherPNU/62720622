```python
import random

answer = random.randint(1, 100)
user = 0
count = 0

while user != answer:
    user = int(input('숫자를 입력하시오: '))
    count = count + 1

    if user < answer:
        print('더 큰 수를 입력하시오')
    elif user > answer:
        print('더 작은 수를 입력하시오')

print(f'정답입니다 {count}회 시행했습니다.')
```

```python
number=int(input('숫자를 입력하시오'))
divisors = [i for i in range(1, number + 1) if number % i == 0]
print(f"{number}의 약수: {divisors}")
```
